# **Project: Personal Reading Journal and Quote Collector**
**You will build a command-line application that helps users log the books they read and collect their favorite quotes, using Python basics, control flow, data structures, functions, string operations, and file handling.
**


# **Requirements**

**Book Entry:**

Each book entry must have:


* Title (string)


* Author (string)


* Year Published (integer)


* Genre(s) (store as a set)


* Status (string: “reading”, “completed”, “on hold”)


**Menu Options:**

* Add a new book entry (unique title + author)


* View all books (sorted by year published)


* Search books by genre or status


* Update a book’s status or add genres


* Delete a book entry


**Quote Collection:**

* Add a new quote (text, book title, page number)


* View all quotes (sorted by book or author)


**Programming Requirements:**

* Use lists, dictionaries, and sets where needed


* Use loops and if-else for input validation


* Write at least three user-defined functions


* All data must be saved to files and loaded at program start


* All string comparisons should be case-insensitive


**Analyzing Features:**

* list all books completed in a user-given year


* Show the book with the most collected quotes


* Find the author(s) with the most entries


```python
books = []
quotes = []

def add_book():
    title = input("Title: ")
    author = input("Author: ")
    year = int(input("Year: "))
    genres = input("Genres (comma separated): ").split(",")
    status = input("Status (reading/completed/on hold): ")

    for book in books:
        if book["title"].lower() == title.lower() and book["author"].lower() == author.lower():
            print("This book already exists.")
            return
    books.append({
        "title": title,
        "author": author,
        "year": year,
        "genres": genres,
        "status": status
    })
    print("Book added.")


def view_books():
     for book in books:
        print(books)


def add_quote():
    text = input("Quote: ")
    book_title = input("Book Title: ")
    page = input("Page Number: ")

    quotes.append({
        "text": text,
        "book": book_title,
        "page": page
    })
    print("Quote added.")

def view_quotes():
        print(quotes)
def search_quotes():
    keyword = input("Enter keyword to search: ").lower()
    found = False
    for q in quotes:
        if keyword in q["text"].lower():
            print(quotes)
            found = True
    if not found:
        print("No quotes found with that keyword.")

def delete_quotes():
    dlt=input("enter quote which u want to dlt? ")
    new_quotes = [q for q in quotes if q.get("text", "").lower() != dlt.lower()]
    if len(new_quotes) < len(quotes):
        quotes[:] = new_quotes
        print("Quote deleted.")
    else:
        print("No quote found with that text.")
def books_completed_in_year():
        year = int(input("Enter year: "))

        found = False
        for book in books:
            if book["year"] == year and book["status"].lower() == "completed":
                print(book)
                found = True

        if not found:
            print("No books completed in that year.")

def book_with_most_quotes():
    if not quotes:
        print("No quotes available.")
        return

    quote_count = {}
    for q in quotes:
        title = q["book"]
        quote_count[title] = quote_count.get(title, 0) + 1

    most_quoted = max(quote_count, key=quote_count.get)
    print("Book with most quotes:", most_quoted)
    print("Number of quotes:", quote_count[most_quoted])

def authors_with_most_entries():
    if not books:
        print("No books in the list.")
        return

    author_count = {}
    for book in books:
        author = book["author"]
        author_count[author] = author_count.get(author, 0) + 1

    max_count = max(author_count.values())
    top_authors = [author for author, count in author_count.items() if count == max_count]

    print("Author(s) with the most entries:")
    for author in top_authors:
        print(author, "-", author_count[author], "book(s)")


def menu():
    while True:
        print("\n1. Add Book")
        print("2. View Books")
        print("3. Add Quote")
        print("4. View Quotes")
        print("5. Search Quotes by Keyword")
        print("6. Delete a Quote")
        print("7. List Books Completed in a Year")
        print("8. Book with Most Quotes")
        print("9. Author(s) with Most Entries")
        print("10. Exit")

        try:
            choice = int(input("Choice: "))
        except ValueError:
            print("Please enter a valid number.")
            continue

        if choice == 1:
            add_book()
        elif choice == 2:
            view_books()
        elif choice == 3:
            add_quote()
        elif choice == 4:
            view_quotes()
        elif choice == 5:
            search_quotes()
        elif choice == 6:
            delete_quotes()
        elif choice == 7:
            books_completed_in_year()
        elif choice == 8:
            book_with_most_quotes()
        elif choice == 9:
            authors_with_most_entries()
        elif choice == 10:
            break
        else:
            print("Invalid choice.")
menu()
```
